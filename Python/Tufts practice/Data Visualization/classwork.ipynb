# **Imports**


```python
import numpy as np
import pandas as pd
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

# **Data Exploration**

```python
datavg = pd.read_csv('data files/vgsales.csv')
names = datavg['Name']
platforms = datavg['Platform']
years = datavg['Year']
genres = datavg['Genre']
publishers = datavg['Publisher']
sales = datavg['Global_Sales']

datag = pd.read_csv('data files/StudentsPerformance.csv')
math = datag['math score']
reading = datag['reading score']
writing = datag['writing score']
people = datag['gender'].count()
math.nlargest(10)
```

# **Data Visualizaion**

```python
datavg.sort_values(by='Year')
```

#**Histogram Graph**

```python
math.hist(color = '#7FDBB6')
plt.xlabel('Score')
plt.ylabel('Number of peope')
plt.title('Score on Math')
plt.grid(False)
```

```python
writing.hist(color = '#7FDBB6')
plt.xlabel('Score')
plt.ylabel('Number of peope')
plt.title('Score on Writing')
plt.grid(False)
```

```python
reading.hist(color = '#7FDBB6')
plt.xlabel('Score')
plt.ylabel('Number of peope')
plt.title('Score on Reading')
plt.grid(False)
```

#**Bar Graph**

```python
datavg_filtered = datavg[(datavg['Year'] >= 1970) & (datavg['Year'] <= 2000)]
datavg_filtered.nlargest(10, 'Global_Sales').plot.barh(y = 'Global_Sales', x = 'Name', color = '#7FDBB6')
```

#**Line Graph**

```python
fig = datavg.sort_values(by = 'Year').plot.line(x = 'Year', y = 'Global_Sales')
fig.grid(axis = 'y')
```

#**Scatter Plot**

```python
fig = datag.sort_values(by = 'math score').nlargest(800,'math score').plot.scatter(x = 'math score', y = 'reading score')
```

#**Pie Chart**

```python
datavg['Publisher'].value_counts().nlargest(20).plot.pie(figsize = (10,10))
plt.title('Top 20 Publishers')
plt.show()
```
